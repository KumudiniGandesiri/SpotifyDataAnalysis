```python
import seaborn as sns
import matplotlib.pyplot as plt

# Print column names to verify existence
print(df_tracks.columns)

# Select only numeric columns and drop specified columns if they exist
numeric_df_tracks = df_tracks.select_dtypes(include=[float, int])

# Drop the specified columns if they exist in the DataFrame
columns_to_drop = ["key", "mode", "explicit"]
numeric_columns_to_drop = [col for col in columns_to_drop if col in numeric_df_tracks.columns]

# Drop specified columns and calculate the Pearson correlation matrix
corr_df = numeric_df_tracks.drop(numeric_columns_to_drop, axis=1).corr(method='pearson')

# Set up the plot figure
plt.figure(figsize=(14, 6))

# Create the heatmap
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidth=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data creation (replace this with your actual data)
data = {
    'release_date': ['2021-01-01', '2020-05-01', '2019-07-10', '2018-12-20', '2021-03-15'],
    'duration': [210, 180, 200, 220, 190]  # duration in seconds
}
df_tracks = pd.DataFrame(data)

# Ensure 'release_date' is in datetime format
df_tracks["release_date"] = pd.to_datetime(df_tracks["release_date"])

# Extract year from 'release_date'
df_tracks["year"] = df_tracks["release_date"].dt.year

# Calculate the average duration per year
average_duration_per_year = df_tracks.groupby("year")["duration"].mean()

# Verify the average duration per year
print(average_duration_per_year)

# Plot line graph using seaborn
sns.set_style("whitegrid")
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x=average_duration_per_year.index, y=average_duration_per_year.values, ax=ax)
ax.set_title("Year vs Average Duration")
ax.set_xlabel("Year")
ax.set_ylabel("Average Duration (seconds)")
plt.xticks(rotation=90)
plt.show()
```
